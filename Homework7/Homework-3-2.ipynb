```python
def avg_score_calc (df_1):
    hotel_names_list = list(set(df_1['Hotel_name']))
    gb = df_1.groupby('Hotel_name')['count']
    gb.count()
    tempt_list = []
    for each_hotel_name in hotel_names_list:
        hotel_str_list = gb.get_group(each_hotel_name).tolist()
        hotel_int_list = [int(x.replace(',','')) for x in hotel_str_list]
        hotel_avg_score = hotel_int_list[0]*5 + hotel_int_list[1]*4 + hotel_int_list[2]*3 + hotel_int_list[3]*2 + hotel_int_list[4]*1
        hotel_avg_score /= sum(hotel_int_list)
        isFiveStar = ((hotel_int_list[0]/sum(hotel_int_list)) >= 0.6)
        tempt_list.append((each_hotel_name, hotel_avg_score, isFiveStar))
        hotel_str_list=[]
        hotel_int_list=[]
    df = pd.DataFrame(tempt_list, columns=['hotel_name','avg_rating','isFiveStar'])
    return df
        
```

```python
df_avg_score = avg_score_calc(df_1)
df_avg_score
```

```python
import pandas as pd
import statsmodels.formula.api as sm

df = pd.DataFrame.from_csv('hotel_linear_regression_rawData.csv')
result = sm.ols(formula="avg_rating ~ Service + Cleanliness + Value + Sleep_Quality + Rooms + Location", data=df).fit()
print(result.summary())
```

Finally, we will use logistic regression to decide if a hotel is _excellent_ or not. We classify a hotel as _excellent_ if more than **60%** of its ratings are 5 stars. This is a binary attribute on which we can fit a logistic regression model. As before, use the model to analyze the data.

```python
import pandas as pd
import statsmodels.formula.api as sm

df = pd.DataFrame.from_csv('hotel_linear_regression_rawData.csv')
df[['isFiveStar']] = df[['isFiveStar']].astype(int)
df.columns = ['hotel_name','avg_rating','Excellence','Value','Rooms','Service','Cleanliness','Location','Sleep_Quality']
df
```

```python
train_cols = df.columns[3:]
logit = sm.Logit(df["Excellence"], df[train_cols])
result = logit.fit()
result.summary()
```

According to the above logistic regression results, we tell that this is pretty much different from the OLS one. The Pseaudo R-squared value is rather low in the above summary plot, and the P-value of the regression is low to be significant, the variables shown less fit in general compared to that of OLS.
